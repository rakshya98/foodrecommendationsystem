```python
ratings_with_name = ratings.merge(food, on='FoodID')
ratings_with_name
```

```python
ratings_with_name.head()
```

```python
ratings_df = ratings_with_name.groupby('FoodName').count()['Ratings'].reset_index()
ratings_df.rename(columns={'Ratings':'num_ratings'}, inplace=True)
ratings_df
```

```python
avg_ratings = ratings_with_name.groupby('FoodName').mean()['Ratings'].reset_index()
avg_ratings.rename(columns={'Ratings':'avg_ratings'}, inplace=True)
```

```python
avg_ratings
```

```python
popular_df = avg_ratings.merge(ratings_df, on='FoodName')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>5].sort_values('avg_ratings', ascending=False)
```

```python
popular_df=popular_df.merge(food,on='FoodName')

popular_df
```

```python
x = ratings_with_name.groupby('UserId').count()['Ratings']>1
foodie_people = x[x].index
foodie_people
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['UserId'].isin(foodie_people)]
filtered_ratings
```

```python
y = filtered_ratings.groupby('FoodName').count()['Ratings']>1
famous_food = y[y].index
famous_food
```

```python
final_ratings = filtered_ratings[filtered_ratings['FoodName'].isin(famous_food)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='FoodName',columns='UserId',values='Ratings')
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(FoodName):
    if FoodName in pt.index:
        index = np.where(pt.index == FoodName)[0][0]
        similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=False)[1:6]
        data = []
        for i in similar_items:
            item = []
            temp_df = food[food['FoodName'] == pt.index[i[0]]]
            item.extend(list(temp_df.drop_duplicates('FoodName')['FoodName'].values))
            item.extend(list(temp_df.drop_duplicates('FoodName')['Category'].values))
            item.extend(list(temp_df.drop_duplicates('FoodName')['Total Calories(Cal)'].values))
            item.extend(list(temp_df.drop_duplicates('FoodName')['Carbs(g)'].values))
            item.extend(list(temp_df.drop_duplicates('FoodName')['Fat(g)'].values))
            item.extend(list(temp_df.drop_duplicates('FoodName')['Protein(g)'].values))
            item.extend(list(temp_df.drop_duplicates('FoodName')['ServingSize'].values))
            item.extend(list(temp_df.drop_duplicates('FoodName')['Image'].values))
            item.extend(list(temp_df.drop_duplicates('FoodName')['Ingredients'].values))
            item.extend(list(temp_df.drop_duplicates('FoodName')['Type'].values))
            item.extend(list(temp_df.drop_duplicates('FoodName')['Veg_NonVeg'].values))
            data.append(item)
        return data
    else:
        return "FoodName not found in recommendations."

result = recommend('Cucumber Pickle')
print(result)
```
